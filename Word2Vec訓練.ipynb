```python
import pandas as pd
from tqdm._tqdm_notebook import tqdm_notebook

import logging
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
from gensim.models.word2vec import Word2Vec

import jieba
jieba.set_dictionary('helper/dict.txt.big')
jieba.initialize()
```

## 刪除null的文本

```python
documents = pd.read_pickle('word2vec/news-corpus.pkl').document.dropna()
print("Shape:", documents.shape)
```

## 文本分詞

```python
%%time
tqdm_notebook.pandas(desc="分詞")
sentences = documents.progress_apply(lambda x: list(jieba.cut(x)))
```

```python
sentences.sample(5)
```

```python
sentences.to_pickle('word2vec/news-sentences.pkl')
```

## 訓練詞向量

```python
%%time
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = Word2Vec(sentences, size=50, sg=1, iter=10, workers=8)
```

```python
model.save('word2vec/d50-sg1-iter10.model')
```

### 批量檢視詞向量的效果

```python
def most_similar(w2v_model, words, topn=10):
    similar_df = pd.DataFrame()
    for word in words:
        try:
            similar_words = pd.DataFrame(w2v_model.wv.most_similar(word, topn=topn), columns=[word, 'cos'])
            similar_df = pd.concat([similar_df, similar_words], axis=1)
        except:
            print(word, "not found in Word2Vec model!")
    return similar_df
```

```python
most_similar(model, ['冬天', '母親', '日本','股票', '車禍', '水果', '三星'])
```

```python
most_similar(model, ['，', '。', '！', '？', '：', '；', '「', '」'])
```
